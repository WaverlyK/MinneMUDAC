```python
import pandas as pd
import numpy as np

# Import all data
july = pd.read_csv('july2020_updated.csv')
trump = pd.read_csv('trump_per_day_with_percent.csv')
soil = pd.read_csv('soil_perday.csv')
weight = pd.read_csv('average_sum_per_day.csv')
sp500 = pd.read_csv('SP500_Soy_Cotton.csv')

# Change date format
july['Date'] =  pd.to_datetime(july['Date'])
trump['Date'] =  pd.to_datetime(trump['Date'])
soil['Date'] =  pd.to_datetime(soil['Date'])
weight['Date'] =  pd.to_datetime(weight['Date'])
sp500['Date'] =  pd.to_datetime(sp500['Date'])

# remove timestamp
july['Date'] = july['Date'].dt.date
trump['Date'] = trump['Date'].dt.date
soil['Date'] = soil['Date'].dt.date
weight['Date'] = weight['Date'].dt.date
sp500['Date'] =  sp500['Date'].dt.date

data = july[['Date','Close']].copy()

## remove unneeded date (uncultured method)

# trump data
remove = []
for i in range(trump.shape[0]):
    flag = 0
    for j in range(data.shape[0]):
        if data.loc[j,'Date']==trump.loc[i,'Date']:
            flag = 1
            break
    if flag == 0:
        remove.append(i)
trump = trump.drop(remove).reset_index(drop=True)

# soil data
remove = []
for i in range(soil.shape[0]):
    flag = 0
    for j in range(data.shape[0]):
        if data.loc[j,'Date']==soil.loc[i,'Date']:
            flag = 1
            break
    if flag == 0:
        remove.append(i)
soil = soil.drop(remove).reset_index(drop=True)

# weight data
remove = []
for i in range(weight.shape[0]):
    flag = 0
    for j in range(data.shape[0]):
        if data.loc[j,'Date']==weight.loc[i,'Date']:
            flag = 1
            break
    if flag == 0:
        remove.append(i)   
weight = weight.drop(remove).reset_index(drop=True)

# SP500 data
remove = []
for i in range(sp500.shape[0]):
    flag = 0
    for j in range(data.shape[0]):
        if data.loc[j,'Date']==sp500.loc[i,'Date']:
            flag = 1
            break
    if flag == 0:
        remove.append(i)   
sp500 = sp500.drop(remove).reset_index(drop=True)
```

```python
# Combine all data into one

# Make sure every row has the same date. If not, raise flag
for i in range(data.shape[0]):
    if data.loc[i,'Date']!=trump.loc[i,'Date']:
        print('trump: ', data.loc[i,'Date'], trump.loc[i,'Date'])
    if data.loc[i,'Date']!=soil.loc[i,'Date']:
        print('soil: ', data.loc[i,'Date'], soil.loc[i,'Date'])
    if data.loc[i,'Date']!=weight.loc[i,'Date']:
        print('weight: ', data.loc[i,'Date'], weight.loc[i,'Date'])
    if data.loc[i,'Date']!=sp500.loc[i,'Date']:
        print('sp500: ', data.loc[i,'Date'], sp500.loc[i,'Date'])

# concat all data
data = pd.concat([data,trump[['Likes','Retweets','percent']],soil[['None','D0','D1','D2','D3','D4']],weight['average_sum'],sp500[['SP500Close','SoyBeanPrice','CottonPrice']]],axis=1)
data.head()
```

```python
print('final july data size: ', data.shape)

data.to_csv('final_data.csv')
```
