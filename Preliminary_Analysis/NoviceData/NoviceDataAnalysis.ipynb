### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.svm import SVC
```

### Functions

```python
def errorRate(ytrue, ypred):
    confusion = np.zeros((2,2))
    for i in range(ytrue.shape[0]):
        confusion[int(ytrue[i]),int(ypred[i])] += 1
        
    return 1 - np.trace(confusion)/ytrue.shape[0]

def normalize(X):
    mean = X.mean()
    std = X.std()
    
    return (X-mean)/std
```

### Import novice data

```python
data = pd.read_csv('noviceData.csv')
data.head()
```

### Consider only US exports

```python
# filter data to US export only
dataUsExport = data.loc[:,:'Agricultural Related 5/']
print('data size: ', dataUsExport.shape)
dataUsExport.head()
```

### Perform Pearson's Correlation to find trend between soybean and the rest

```python
correlation = []
for i in range(2,m):
    if dataColumn[i] != 'Soybeans':
        correlation.append(stats.pearsonr(dataUsExport.loc[:,'Soybeans'],dataUsExport.iloc[:,i]))
    
correlation = pd.DataFrame(correlation, index = np.delete(dataColumn,[0,1,8]))
correlation 
```

### Feature in which Pearson Correlation larger than 0.9

```python
selectedFeatures = correlation[correlation.iloc[:,0]>0.9]
selectedFeatures
```

### Perform simple classification using the features above

```python
selectedData = dataUsExport[np.append(selectedFeatures.index.values, 'Soybeans')]
monthlyPercentageChange = []
for i in range(1,selectedData.shape[0]):
    monthlyPercentageChange.append((selectedData.iloc[i,:] - selectedData.iloc[i-1,:])/selectedData.iloc[i-1,:])

monthlyPercentageChange = np.array(monthlyPercentageChange)
outputData = monthlyPercentageChange[:,-1]

# normalize X
X = normalize(np.delete(monthlyPercentageChange, monthlyPercentageChange.shape[0]-1, axis = 0)*100)
y = np.sign(np.delete(outputData, 0, axis = 0))
y[y==0]=1 #set 0 to 1
y[y==-1]=0 #set -1 to 0

Xtrn = X[:350,:]
Xval = X[350:400,:]
Xtst = X[400:,:]
ytrn = y[:350]
yval = y[350:400]
ytst = y[400:]

print("Xtrn: ", Xtrn.shape, " ytrn: ", ytrn.shape, "Xval: ", Xval.shape, " yval: ", yval.shape, "Xtst: ", Xtst.shape, " ytst: ", ytst.shape)
```

### Linear SVM Classifier

```python
Cs = [1e-4, 1e-3, 1e-2, 1e-1, 1e-0, 1e1, 1e2, 1e3, 1e4]

trnError = []
valError = []

for C in Cs:
    model = SVC(C=C, kernel='linear')
    model.fit(Xtrn,ytrn)
    
    ypred = model.predict(Xtrn)
    trnError.append(errorRate(ytrn, ypred))
    
    ypred = model.predict(Xval)
    valError.append(errorRate(yval, ypred))

    
# Train Optimal model    
opt_C = Cs[np.argmin(valError)]
opt_model = SVC(C=opt_C, kernel='linear')
model.fit(np.append(Xtrn,Xval,axis=0),np.append(ytrn,yval,axis=0)) 
ypred = model.predict(Xtst)
testError = errorRate(ytst, ypred)
print('Opt C: ', opt_C, ' test error: ', testError)    
```

### RBF SVM Classifier

```python
Cs = [1e-4, 1e-3, 1e-2, 1e-1, 1e-0, 1e1, 1e2, 1e3, 1e4, 1e5]
gammas = [1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3]

trnError = np.zeros((len(Cs), len(gammas)))
valError = np.zeros((len(Cs), len(gammas)))

gcount = 0
for gamma in gammas:
    ccount = 0
    for C in Cs:
        model = SVC(C=C, gamma = gamma, kernel='rbf')
        model.fit(Xtrn,ytrn)

        ypred = model.predict(Xtrn)
        trnError[ccount,gcount] = errorRate(ytrn, ypred)

        ypred = model.predict(Xval)
        valError[ccount,gcount] = errorRate(yval, ypred)
        
        ccount += 1
    gcount += 1
        
    
# Train Optimal model    
opt_C = Cs[np.argmin(valError.min(axis=1))]
opt_gamma = gammas[np.argmin(valError.min(axis=0))]
opt_model = SVC(C=opt_C, gamma=opt_gamma, kernel='rbf')
model.fit(np.append(Xtrn,Xval,axis=0),np.append(ytrn,yval,axis=0)) 
ypred = model.predict(Xtst)
testError = errorRate(ytst, ypred)
print('Opt C: ', opt_C, ' opt gamma : ', opt_gamma, ' test error: ', testError) 
```
