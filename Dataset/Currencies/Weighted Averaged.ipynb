```python
import pandas as pd

# currency price data
datasetNames = ['USD_ARS Historical Data.csv','USD_BRL Historical Data.csv',
              'USD_CAD Historical Data.csv','USD_CNY Historical Data.csv','USD_INR Historical Data.csv',
              'USD_PYG Historical Data.csv']

# Production per year per country
production = [[56.8,57.8,37.8,55],[96.5,114,120.8,116.5],[6.24,6.46,7.72,7.3],
             [11.6,12.9,15.2,15.9],[7,11.5,8.35,11],[9,10.67,10.3,9]]

data = pd.read_csv(datasetNames[0])
data = pd.to_datetime(data['Date'])

for name in datasetNames:
    price = pd.read_csv(name)
    price = price['Price'].rename("Price_%s" % (name[4:7]))
    if name == 'USD_PYG Historical Data.csv':
        for i in range(price.shape[0]):
            price[i] = price[i].replace(',','')
    
    price = 1/pd.to_numeric(price)
    data = pd.concat([data,price],axis=1)

data2 = data.copy()
for i in range(data2.shape[0]):
    year = data2.loc[i,'Date'].year
    
    if year == 2019:
        idx = 3
    elif year == 2018:
        idx = 2
    elif year == 2017:
        idx = 1
    else:
        idx = 0
        
    for j in range(6):
        data2.iloc[i,j+1] /= production[j][idx]

summation = data2.iloc[:,1:].sum(axis=1).rename('average_sum')

dataNew = pd.concat([data2['Date'],summation],axis=1)

dataNew.to_csv('average_sum.csv')

dataNew.head()
```
